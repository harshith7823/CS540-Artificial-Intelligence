```python
import numpy as np
from collections import defaultdict
import heapq as hq
```

```python
def manhattan_dist(a):
  a = np.reshape(a, (3,3))
  actual = np.array([1,2,3,4,5,6,7,8,0])
  actual = np.reshape(actual, (3,3))
  manhattan_dist = 0
  for i in range(1, 9):
    should_be_pos = np.argwhere(actual==i)[0]
    actual_pos = np.argwhere(a==i)[0]
    manhattan_dist += ( abs(should_be_pos[0] - actual_pos[0]) + abs(should_be_pos[1] - actual_pos[1]))
  return manhattan_dist

a = np.array([1, 0, 2, 3, 4, 5, 6, 7, 8])
a = np.reshape(a, (3,3))
print(manhattan_dist(a))
```

```python
def swap_positions(a):
    # rename stuff here
    row, col = np.argwhere(a==0)[0]
    free = []
    if row > 0:
      free.append((row - 1, col))
    if col > 0:
      free.append((row, col - 1))
    if row < 2:
      free.append((row + 1, col))
    if col < 2:
      free.append((row, col + 1))
    return free
```

```python
def print_succ(a):
  
  a = np.reshape(a, (3,3))
  b = np.argwhere(a==0)[0]
  pos = swap_positions(a)

  result = []
  for move in pos:
    c = np.copy(a)
    c[move],c[tuple(b)] =  c[tuple(b)], c[move]
    p = c.flatten().tolist()
    result.append((p, manhattan_dist(p)))
  
  res = sorted(result)
  # for i,j in res:
  #   print(i, "h=" + str(j))
  

  return res

a = [1,2,3,0,4,5,6,7,8]
print("Successor states ->")
succ = print_succ(a)
```

```python

nodes = defaultdict(None)

visited = set()

# s = [4,3,8,5,1,6,7,2,0]
s =  [1,2,3,4,5,6,7,0,8]
g = 0
h = manhattan_dist(s)

pq = []
hq.heappush(pq,(g+h, s, (g, h, -1, 0)))

visited.add(s)

n = 0
i = 0
maxsize = 1
nodes[i] = s

ans = -1

while len(pq):

  expand = list(hq.heappop(pq))
  # d[i] = expand
  # i+=1
  if expand[2][1] == 0:
    print("reached goal state")
    ans = expand
    break

  # print("expanding")
  # print(expand)
  # print("--")
  succ = print_succ(expand[1])

  for state, h in succ:
    if state not in visited:
      visited.add(state)
      g = expand[2][0] + 1
      hq.heappush(pq,(g+h, state, (g, h, expand[2][3], i)))
      nodes[i] = (state,  expand[2][3])
      i += 1
  
  maxsize = max(maxsize, len(pq))

  n += 1 


cur = ans[2][3]

while nodes[cur][1] != -1:
  print(nodes[cur][0])
  cur = nodes[cur][1]







```
