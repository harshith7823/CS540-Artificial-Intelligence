```python
from scipy.linalg import eigh   
import numpy as np  
import matplotlib.pyplot as plt 
from scipy.linalg import block_diag
```

```python
def load_and_center_dataset(filename):
  x = np.load(filename)   
  mean = np.mean(x, axis=0)
  return x - mean
```

```python
def get_covariance(dataset):
  return np.dot(np.transpose(dataset),dataset)/(len(dataset)-1)
```

```python
def get_eig(S, m):
  rows, cols = S.shape
  eigen_val, eigen_vect = eigh(S, subset_by_index=[rows-m, rows-1])
  eigen_val = np.flip(eigen_val)
  eigen_vect = np.flip(eigen_vect, axis=1)
  diag_matrix = np.diag(eigen_val)
  return diag_matrix, eigen_vect
```

```python
def get_eig_perc(S, perc):
  eigen_val, eigen_vect = eigh(S, subset_by_value=[perc*np.trace(S), np.trace(S)])
  eigen_val = np.flip(eigen_val)
  eigen_vect = np.flip(eigen_vect, axis=1)
  diag_matrix = np.diag(eigen_val)
  return diag_matrix, eigen_vect
```

```python
def project_image(img, U):
  alpha = U.T.dot(img)
  num_eigen_vect = U.shape[1]
  len_eigen_vect = U.shape[0]
  proj = np.zeros(len_eigen_vect,)  
  for i in range(num_eigen_vect):
    proj += alpha[i] * U[:,i]
  return proj
```

```python
def display_image(orig, proj):
  f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
  ax1.set_title('Original')
  pos1 = ax1.imshow(orig.reshape(32,-1).T, aspect='equal')
  f.colorbar(pos1, ax=ax1)

  ax2.set_title('Projection')
  pos2 = ax2.imshow(proj.reshape(32,-1).T, aspect='equal') 
  f.colorbar(pos2, ax=ax2)
  plt.show()
```
